## Multivariate Distributions in Python

Sometimes we can get a lot of information about how two variables (or more) relate if we plot them together. This tutorial aims to show how plotting two variables together can give us information that plotting each one separately may miss.



```python
%matplotlib inline
# Inline Plotting: When you use %matplotlib inline, it ensures that all matplotlib plots are rendered directly within the notebook itself, right below the code cell that generates them. This means you will see the visualizations immediately in the notebook, which is particularly useful for interactive data analysis and exploration.
#
# Persistent Display: It makes the plots persistent within the notebook's output cells, so you don't need to use plt.show() explicitly to render the plots.

# import the packages we are going to be using
import numpy as np # для работы с массивами и генерации данных
import matplotlib.pyplot as plt # для создания графиков
import seaborn as sns; sns.set() # для использования другой темы для графиков

# Don't worry so much about what rho is doing here
# Just know if we have a rho of 1 then we will get a perfectly
# upward sloping line, and if we have a rho of -1, we will get 
# a perfectly downward slopping line. A rho of 0 will 
# get us a 'cloud' of points

# Настройка параметра корреляции
# r = 0.5
# r: Это коэффициент корреляции, который определяет степень связи между двумя переменными.
# Значение 0.5 указывает на умеренную положительную корреляцию.

r = 0.5

# Don't worry so much about the following three lines of code for now
# this is just getting the data for us to plot
mean = [15, 5]
cov = [[1, r], [r, 1]]

# mean: Список, содержащий средние значения для двух переменных (по умолчанию 15 и 5).

# cov: Матрица ковариации, которая описывает степень корреляции между переменными.
# В данном случае, [[1, r], [r, 1]] задает ковариацию переменных с корреляцией r.

# np.random.multivariate_normal: Генерирует 400 случайных точек на основе заданного среднего и ковариации.
# x и y содержат сгенерированные данные для каждой переменной.

x, y = np.random.multivariate_normal(mean, cov, 400).T

# Adjust the figure size
plt.figure(figsize=(10,5)) # Определяет размер графика: ширина 10 дюймов и высота 5 дюймов.

# Plot the histograms of X and Y next to each other

plt.subplot(1,2,1) # Создает первую подграфику (гистограмму) в сетке из одной строки и двух столбцов.
plt.hist(x = x, bins = 15) # Создает гистограмму для переменной x с 15 корзинами (бинами).
plt.title("X") # Задает заголовок для первой подграфики.

plt.subplot(1,2,2) #  Создает вторую подграфику (гистограмму) в той же сетке.
plt.hist(x = y, bins = 15) # Создает гистограмму для переменной y с 15 корзинами (бинами).
plt.title("Y") # Задает заголовок для второй подграфики.

plt.show()
```

```python
# Plot the data
plt.figure(figsize=(10,10))
plt.subplot(2,2,2)
plt.scatter(x = x, y = y)
plt.title("Joint Distribution of X and Y")

# Plot the Marginal X Distribution
plt.subplot(2,2,4)
plt.hist(x = x, bins = 15)
plt.title("Marginal Distribution of X")

# Plot the Marginal Y Distribution
plt.subplot(2,2,1)
plt.hist(x = y, orientation = "horizontal", bins = 15)
plt.title("Marginal Distribution of Y")

# Show the plots
plt.show()
```
