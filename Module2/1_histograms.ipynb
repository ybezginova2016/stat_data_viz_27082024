```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
################################### CREATING HISTOGRAM ###################################
#
# Данные: В этом примере данные heights представляют собой рост взрослых мужчин в дюймах. Вы можете заменить эти данные на свои.

# Построение гистограммы: Используется функция plt.hist() для создания гистограммы. Параметр bins=8 задает количество интервалов (столбцов) на гистограмме.
#
# Линия плотности: Линия плотности добавлена для наглядности и используется метод gaussian_kde из библиотеки scipy.
#
# Оформление: Заголовки и подписи к осям добавляются с помощью plt.title(), plt.xlabel(), и plt.ylabel().
#
# Показ графика: График отображается с помощью plt.show().

################################### END ###################################
```

```python
# Пример данных о росте взрослых мужчин в дюймах
heights = np.array([62, 64, 65, 67, 68, 68, 69, 70, 71, 72, 74, 67, 66, 70, 72, 68, 69, 71, 70, 73])

# Построение гистограммы
plt.hist(heights, bins=8, color='blue', edgecolor='black', alpha=0.7)

# Добавление линии плотности (опционально, для наглядности)
# Для этого используем гауссовское сглаживание (Kernel Density Estimate)
from scipy.stats import gaussian_kde
density = gaussian_kde(heights)
xs = np.linspace(min(heights), max(heights), 200)
plt.plot(xs, density(xs)*len(heights), color='red')

# Настройка заголовков и подписей
plt.title('Histogram of Adult Male Heights')
plt.xlabel('Height (Inches)')
plt.ylabel('Frequency')

# Показ гистограммы
plt.show()
```
